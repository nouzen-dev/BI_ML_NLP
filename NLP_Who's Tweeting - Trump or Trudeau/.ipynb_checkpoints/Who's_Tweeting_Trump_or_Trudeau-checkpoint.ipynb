# Tweet classification: Trump vs. Trudeau

<img src="https://upload.wikimedia.org/wikipedia/commons/4/47/President_Donald_Trump_and_Prime_Minister_Justin_Trudeau_Joint_Press_Conference%2C_February_13%2C_2017.jpg" title="Donald Trump & Justin Trudeau"/>

Problem: Can we classify tweets based on the author?<br>
Task: Build a classifier to classify tweets from two prominent North American politicians: Donald Trump and Justin Trudeau.

```python
import random; random.seed(53)

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn import metrics
```

## Transforming our collected data
A corpus of tweets in csv format collected via Twitter API

```python
import pandas as pd

tweet_df = pd.read_csv('datasets/tweets.csv')

y = tweet_df['author']

X_train, X_test, y_train, y_test = train_test_split(tweet_df['status'], y, random_state = 53, test_size = 0.33)
```

## Vectorize tweets

```python
count_vectorizer = CountVectorizer(stop_words = 'english', min_df= 0.05, max_df = 0.9)

count_train = count_vectorizer.fit_transform(X_train)
count_test = count_vectorizer.transform(X_test)

tfidf_vectorizer = TfidfVectorizer(stop_words = 'english', min_df = 0.05, max_df = 0.9)

tfidf_train = tfidf_vectorizer.fit_transform(X_train)
tfidf_test = tfidf_vectorizer.transform(X_test)
```

## Training a multinomial naive Bayes model

Investigate using the Multinomial Naive Bayes model with both the CountVectorizer and TfidfVectorizer data to see which performs better by assessing test sets accuracy scores for both models

```python
tfidf_nb = MultinomialNB()
tfidf_nb.fit(tfidf_train, y_train)

tfidf_nb_pred = tfidf_nb.predict(tfidf_test)

tfidf_nb_score = metrics.accuracy_score(y_test, tfidf_nb_pred)

count_nb = MultinomialNB()
count_nb.fit(count_train, y_train)

count_nb_pred = count_nb.predict(count_test)

count_nb_score = metrics.accuracy_score(y_test, count_nb_pred)

print('NaiveBayes Tfidf Score: ', tfidf_nb_score)
print('NaiveBayes Count Score: ', count_nb_score)
```

## Evaluating model

The TF-IDF model performs better than the count-based approach, further evaluation of the model is shown using a confusion matrix

```python
%matplotlib inline

from datasets.helper_functions import plot_confusion_matrix

tfidf_nb_cm = metrics.confusion_matrix(y_test, tfidf_nb_pred, labels=['Donald J. Trump', 'Justin Trudeau'])
count_nb_cm = metrics.confusion_matrix(y_test, count_nb_pred, labels=['Donald J. Trump', 'Justin Trudeau'])

plot_confusion_matrix(tfidf_nb_cm, classes=['Donald J. Trump', 'Justin Trudeau'], title="TF-IDF NB Confusion Matrix")

plot_confusion_matrix(count_nb_cm, classes=['Donald J. Trump', 'Justin Trudeau'], title="Count NB Confusion Matrix", figure=1)
```

The Bayesian model only has 1 prediction difference between the TF-IDF and count vectorizers, and there is some confusion when the predicted label is Trump but the actual tweeter is Trudeau. We will use  LinearSVC with the TF-IDF vectors to see if this improves the accuracy of the classifier

```python
tfidf_svc = LinearSVC()
tfidf_svc.fit(tfidf_train, y_train)

tfidf_svc_pred = tfidf_svc.predict(tfidf_test)

tfidf_svc_score = metrics.accuracy_score(y_test, tfidf_svc_pred)

print("LinearSVC Score:   %0.3f" % tfidf_svc_score)

svc_cm = metrics.confusion_matrix(y_test, tfidf_svc_pred, labels=['Donald J. Trump', 'Justin Trudeau'])

plot_confusion_matrix(svc_cm, classes=['Donald J. Trump', 'Justin Trudeau'], title="TF-IDF LinearSVC Confusion Matrix")
```

The LinearSVC model is even better than the Multinomial Bayesian one. Via the confusion matrix, we can see that, although there is still some confusion where Trudeau's tweets are classified as Trump's, the False Positive rate is better than the previous model. WIth the LinearSVC model, we will see what are the most Trump-like or Trudeau-like words
