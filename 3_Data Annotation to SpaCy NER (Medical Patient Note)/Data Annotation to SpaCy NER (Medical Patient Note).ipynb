The text data is from a Kaggle competition (https://www.kaggle.com/competitions/nbme-score-clinical-patient-notes/overview/description), from the USMLE Step 2 Clinical Skills examination, a medical licensure exam. It measures a trainee's ability to recognize pertinent clinical facts during encounters with standardized patients:
1) During this exam, each test taker sees a Standardized Patient (a person trained to portray a clinical case)<br>
2) The test taker documents the relevant facts of the encounter in a patient note. <br>
3) Each patient note is scored by a trained physician who looks for the presence of certain features relevant to the case as described in a rubric.<br><br>
The goal of this competition is to develop an automated way of identifying the relevant features within each patient note, with a special focus on the patient history portions of the notes where the information from the interview with the standardized patient is documented.

Datasets:
- patient_notes- a collection of about 40,000 Patient Note history portions. Only a subset of these have features annotated. The patient notes in the test set are not included in the public version of this file.
- features - the rubric of features (or key concepts) for each clinical case.
- train - feature annotations for 1000 of the patient notes, 100 for each of 10 cases.

## Data Annotation - Programmatically

Manual annotation is a time-consuming and collaborative team effort. For demo, I will try to annotate programatically:

```python
pn_dict = {}
for idx, row in patient_notes.iterrows():
    pn_dict[row['pn_num']] = row['pn_history']
print(len(pn_dict))
pn_dict
```

```python
new_annotation = []

for case_id in features['case_num'].unique():
    all_pn_id = set(patient_notes[patient_notes['case_num']==case_id]['pn_num'].tolist())
    
    for feature_id in features[features['case_num']==case_id]['feature_num'].unique():
        annotated_pn = set(train[train['feature_num']==feature_id]['pn_num'].tolist())  # get pn_num that have been annotated
        pn_to_annotate = all_pn_id - annotated_pn # get pn_num that have NOT been annotated
        
        annotations = train[train['feature_num']==feature_id]['annotation'].tolist() # get all current annotations
        annotation_texts = set()
        for a in annotations:
            anns = eval(a)
            for at in anns:
                annotation_texts.add(at)
                
        # annotate       
        for pn_id in pn_to_annotate:
            new_annotation_pn, new_location_pn = [], []
            pn_text = pn_dict[pn_id]
            for at in annotation_texts:
                start = pn_text.find(at)
                if start>=0:
                    new_annotation_pn.append(at)
                    new_location_pn.append(f'{start} {start+len(at)}')
            if len(new_annotation_pn) > 0:
                new_annotation.append((f'{pn_id:04d}_{feature_id:03d}', case_id, pn_id, feature_id, new_annotation_pn, \
                                       new_location_pn))
        break
```

```python
def plot_50_most_common_words(count_data, count_vectorizer):
    count_vectorizer = CountVectorizer(ngram_range=(3,3), stop_words='english')
    count_data = count_vectorizer.fit_transform(features['features_text_sub'])
    vocab = count_vectorizer.vocabulary_ 
    new = pd.DataFrame.from_dict(vocab, orient='index').reset_index()
    new_sorted = new.sort_values(0, ascending = True)
    new_sorted[:50].plot(kind='barh', x='index', figsize=(15,15), xlabel='Tokens', ylabel='Count', colormap='viridis', \
                     table=False, grid=True, fontsize=12, rot=0, position=1, title='Top 50 Common Set of Words in Features (Trigram)', \
                     legend=True).legend(['Tokens'], loc='lower left', prop={'size':15})
    
plot_50_most_common_words(count_data, count_vectorizer)
```

## We can look at individual patient's annotations & features together:

In the respective patient note, what has been annotated?

```python
def patient_note_returned(pn_idx):
    
    subset = train_appended[train_appended['pn_num']==pn_idx]
    subset
    location  = patient_train['location_sub_split']
    annotation = patient_train["annotation_sub"]
    start_pos = []
    end_pos = []
    ents = []
    feature_list_pn = subset['feature_num'].tolist()

    for i in location:
        for idx, element in enumerate(i):
            if idx % 2 == 0:
                start_pos.append(element)
            else:
                end_pos.append(element)
            
    for i in range(len(start_pos)):
        ents.append({'start': int(start_pos[i]), 'end': int(end_pos[i]), 'label': 'Annotation'})

    print('Patient', pn_idx, '- annotated patient note:')
    text = {'text': patient_note, 'ents': ents}
    colors = {'Annotation': 'linear-gradient(90deg, #FFDF00, #FF8674)'} 
    options = {'colors': colors}
    print(len(ents), "annotations")
    displacy.render(text, style='ent', options=options , manual=True, jupyter=True)
    
    print('Patient', pn_idx, '- features:')
    for feature_num in feature_list_pn:
        feature = features[features['feature_num'] == feature_num]['features_text_sub'].unique()
        print(feature)
    

patient_note_returned(74087)
```
