## N-gram, TFIDF & Cosine Similarity

```python
import nltk
import re

from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")

def tokenize_and_stem(text):
    tokens = [word.lower() for sent in nltk.sent_tokenize(text) for word in nltk.word_tokenize(sent)]
    filtered_tokens = [token for token in tokens if re.search('[a-zA-Z]', token)]
    stems = [stemmer.stem(t) for t in filtered_tokens]
    return stems
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS
tfidf = TfidfVectorizer(stop_words=ENGLISH_STOP_WORDS, tokenizer=tokenize_and_stem, ngram_range=(1,3))
tfidf_matrix = tfidf.fit_transform(x for x in short_movies['description'])
tfidf_matrix.shape
```

```python
short_movies.shape
```

```python
tfidf_print = pd.DataFrame(tfidf_matrix.toarray(), columns = tfidf.get_feature_names_out())
```

```python
tfidf_print.head()
```

```python
tfidf_print.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

def get_recommendations(title, cosine_sim=cosine_sim):
    try:
        idx = short_movies[short_movies['title'] == title].index[0]

        # Get the pairwsie similarity scores of all movies with that movie
        sim_scores = list(enumerate(cosine_sim[idx]))

        # Sort the movies based on the similarity scores
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        
        sim_scores = sim_scores[1:10]

        all_indices = [i[0] for i in sim_scores]

        print('Top 10 recommended 1-hour movies for [{}] are:'.format(title))
     
        return short_movies[['title', 'type', 'duration']].loc[all_indices]
    
    except:
        print('Movie not found')
```
