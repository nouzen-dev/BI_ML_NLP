I will compute the Jaccard Similarity Score for 'text' & 'selected_text' (https://www.geeksforgeeks.org/find-the-jaccard-index-and-jaccard-distance-between-the-two-given-sets/) & look at its distribution across sentiments.<br>

Jaccard Score is a measure of how similar/ dissimilar 2 sets are - the higher the score, the more similar the 2 strings are.

```python
def jaccard(str1, str2): 
    a = set(str1.lower().split()) 
    b = set(str2.lower().split())
    c = a.intersection(b)
    return float(len(c))/(len(a)+len(b)-len(c))
```

```python
results_jaccard=[]

for ind, row in train.iterrows():
    sentence1 = row.text
    sentence2 = row.selected_text
    jaccard_score = jaccard(sentence1,sentence2)
    results_jaccard.append([sentence1,sentence2,jaccard_score])

jaccard = pd.DataFrame(results_jaccard,columns=["text","selected_text","jaccard_score"])
train = train.merge(jaccard,how='outer')
train.head()
```

# Data Preprocessing

As my target column is 'selected_text', all preprocessings will be performed on this column.

As Twitter has character limit, it's natural for contractions to be used. I will first expand the contracted words as this might bring more insights in downstream processing

### Expand Contraction

```python
def decontracted(text):  
    text = re.sub(r"n\'t", " not", text)
    text = re.sub(r"n\t", " not", text)
    text = re.sub(r"n\`t", " not", text)
    text = re.sub(r"\'re", " are", text)
    text = re.sub(r"\re", " are", text)
    text = re.sub(r"\`re", " are", text)
    text = re.sub(r"\'s", " is", text)
    text = re.sub(r"\s", " is", text)
    text = re.sub(r"\`s", " is", text)
    text = re.sub(r"\'d", " would", text)
    text = re.sub(r"\d", " would", text)
    text = re.sub(r"\`d", " would", text)
    text = re.sub(r"\'ll", " will", text)
    text = re.sub(r"\ll", " will", text)
    text = re.sub(r"\`ll", " will", text)
    text = re.sub(r"\'t", " not", text)
    text = re.sub(r"\t", " not", text)
    text = re.sub(r"\`t", " not", text)
    text = re.sub(r"\'ve", " have", text)
    text = re.sub(r"\ve", " have", text)
    text = re.sub(r"\`ve", " have", text)
    text = re.sub(r"\'m", " am", text)
    text = re.sub(r"\m", " am", text)
    text = re.sub(r"\`m", " am", text)
    return text
```

```python
train['selected_text'] = train['selected_text'].apply(lambda x:decontracted(x))
train.head()
```

### Data Cleaning:

```python
# Python comes built-in with a library, string, which includes an attribute string.punctuation that includes many built-in punctuation characters. Because the library is built-in, you dont need to worry about needing to install it.
print(string.punctuation)
print(('[%s]' % re.escape(string.punctuation)))
```

```python
def preprocess(text):
    text = str(text).lower()
    text = re.sub(r"[!\"#\$%&\'\(\)\+,-\./:;<=>\?@\[\\\]\^_`{\|}~]", ' ', text) # remove specific punctuation (retain * & possible emoji)
    text = re.sub('[0-9]+', ' ', text) # remove number
    text = re.sub('(www.[^s]+)', ' ', text) # remove URL
    text = re.sub('(http.?:?[^s]+)', ' ', text) # remove URL (http & https)
    text = re.sub(r'\<.*\>', ' ', text) # remove word in < >
    text = re.sub(r'\[.*\]', ' ', text) # remove text in [ ]
    text = re.sub('\n', ' ', text) # remove new line
    text = re.sub('@[^\s]+', ' ', text) # remove handle
    text = re.sub('#[^\s]+', ' ', text) # remove hashtag
    text = re.sub(r'[^\x00-\x7f]',' ', text) # remove non-ASCII characters 
    return text
```

I will also be performing some simple typo corrections for a few commonly used words & also remove repeating character:

```python
def preprocess_typo(text):
    text = re.sub(r'\btheyre\b', 'they are', text)
    text = re.sub(r'\bgooood\b', 'good', text) 
    text = re.sub(r'\bdont\b', 'not', text)
    text = re.sub(r'\bcannot\b', 'not', text)
    text = re.sub(r'\bdoesnt\b', 'not', text)
    text = re.sub(r'\bshudnt\b', 'not', text)
    text = re.sub(r'\bicant\b', 'not', text)
    text = re.sub(r'\booh\b', 'oh', text)
    text = re.sub(r'\bhapp\b', 'happy', text)
    text = re.sub(r'\blols\b', 'lol', text) 
    text = re.sub(r'\blolz\b', 'lol', text)
    text = re.sub(r'\bsry\b', 'sorry', text)
    text = re.sub(r'\bsorr\b', 'sorry', text)
    text = re.sub(r'\bsrry\b', 'sorry', text)
    text = re.sub(r'\brly\b', 'really', text)
    text = re.sub(r'\bsoo\b', 'so', text)
    text = re.sub(r'\bsooo\b', 'so', text)
    text = re.sub(r'\bsoooo\b', 'so', text)
    text = re.sub(r'\bsooooo\b', 'so', text)
    text = re.sub(r'\bsoooooo\b', 'so', text)
    text = re.sub(r'\bnoooo\b', 'no', text)
    text = re.sub(r'\bluv\b', 'love', text)
    text = re.sub(r'\bluve\b', 'love', text)
    text = re.sub(r'\blov\b', 'love', text)
    text = re.sub(r'\bloveeee\b', 'love', text)
    text = re.sub(r'\bloveeeee\b', 'love', text)
    text = re.sub(r'\bloveeeeeeeeeeeeeeeeeeeeee\b', 'love', text)
    text = re.sub(r'\bloooove\b', 'love', text)
    text = re.sub(r'\blovelovelove\b', 'love', text) 
    text = re.sub(r'\blooove\b', 'love', text)
    text = re.sub(r'\bca\b', 'can', text)
    text = re.sub(r'\bthnks\b', 'thank', text)
    text = re.sub(r'\bthnxs\b', 'thank', text)
    text = re.sub(r'\bthnx\b', 'thank', text)
    text = re.sub(r'\bthankx\b', 'thank', text)
    text = re.sub(r'\byummmm\b', 'yummy', text)
    text = re.sub(r'\bfuunn\b', 'fun', text)
    text = re.sub(r'\byaayy\b', 'yay', text)
    text = re.sub(r'\byeay\b', 'yay', text)
    text = re.sub(r'\bcongradts\b', 'congratulations', text)
    text = re.sub(r'\bdayyyy\b', 'day', text)
    text = re.sub(r'\bgoood\b', 'day', text)
    text = re.sub(r'\bmis\b', 'miss', text)
    text = re.sub(r'\bmothers\b', 'mother', text)
    text = re.sub(r'\bamazin\b', 'amaze', text)
    text = re.sub(r'\bamazi\b', 'amaze', text)
    return (text)
```

### Lemmatization:

```python
def lemmatize_text(text):
    txt = []
    lemmatizer = nltk.stem.WordNetLemmatizer()
    txt = "".join([lemmatizer.lemmatize(w, pos='v') for w in text]) # remember to join the words back
    return txt
```

```python
train['selected_text'] = train['selected_text'].apply(lambda x:preprocess(x))
train['selected_text'] = train['selected_text'].apply(lambda x:lemmatize_text(x))
train['selected_text'] = train['selected_text'].apply(lambda x:preprocess_typo(x))
train.head()
```

```python
train['selected_text'] = train['selected_text'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>1]))
train.head()
```

### Tokenization & Bag-of-words:

```python
def tokenize(text):
    tokens = word_tokenize(text)
    return tokens

def get_top_n_gram(text,ngram_range,n=None):
    vec = CountVectorizer(ngram_range=ngram_range,tokenizer=tokenize,strip_accents='ascii', \
                          preprocessor=preprocess_typo,stop_words=stopwordlist).fit(text) 
                          # remove token_pattern
    bag_of_words = vec.transform(text)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
```

```python
train['selected_text'] = train['selected_text'].apply(lambda x:str(x).split())

raw_text = [word for word_list in train2['selected_text'] for word in word_list] # takes in words not in allother = specificonly
def words_unique(sentiment, numwords, raw_words):
    '''
    Input:
        segment - Segment category (ex. 'Neutral');
        numwords - how many specific words do you want to see in the final result; 
        raw_words - list  for item in train_data[train_data.segments == segments]['selected_text_p']:
    Output: 
        dataframe giving information about the name of the specific ingredient and how many times it occurs in the chosen cuisine (in descending order based on their counts)..

    '''
    allother = []
    mycounter = Counter()
    
    for item in train[train.sentiment!=sentiment]['selected_text']:
        for word in item:
            allother.append(word)
    allother = list(set(allother))
    
    specificnonly = [x for x in raw_text if x not in allother] 
    # = words not in raw_text after comparing with allother
    keep = list(specificnonly)
    
    for item in train2[train2.sentiment==sentiment]['selected_text']: 
        # = words used in the selected sentiment's tweets = not unique words
        for word in item:
            mycounter[word] += 1

    for word in list(mycounter):
        if word not in keep: 
            del mycounter[word]
    # Using the keep list as reference, if the word in counter is not found in the list -> delete it from the counter
    unique_words = pd.DataFrame(mycounter.most_common(numwords), columns=['word','count'])
    print(len(mycounter))
    return unique_words
```

# Prediction of Extracted Text with Transformer BERT with Pytorch

The portion of a tweet that represents the sentiment is the 'selected_text'. The goal of this project is to predict the 'selected_text' based on the 'text' column. I will be using a variant of BERT model, Distilbert, with Pytorch, by framing this problem as a question-answering task. I will use a pretrained model from HuggingFace with simpletransformer (https://simpletransformers.ai/docs/qa-model/), train it with training tweets & predict the 'selected_text' in the 'submission' file.<br><br>
Summary - given a question ('sentiment') + a context ('text'), the transformer model trained will find the answer ('selected_text') from the 'text' column.<br>

The QuestionAnsweringModel in Simpletransformer is chosen as it uses lesser resources. The Distilbert model is chosen because it has been finetuned on question-answering task using the SQuAD dataset.

## Data Preparation in the Q&A Format

### Training Data Preparation:

```python
%time

def find_all(input_str, search_str):
    l1 = []
    length = len(input_str)
    index = 0
    while index < length:
        i = input_str.find(search_str, index)
        if i == -1:
            return l1
        l1.append(i)
        index = i + 1
    return l1

def do_qa_train(train):
    output = []
    for line in train:
        context = line[1]
        qas = []
        question = line[-1]
        qid = line[0]
        answers = []
        answer = line[2]
        if type(answer) != str or type(context) != str or type(question) != str:
            print(context, type(context))
            print(answer, type(answer))
            print(question, type(question))
            continue
        answer_starts = find_all(context, answer)
        for answer_start in answer_starts:
            answers.append({'answer_start': answer_start, 'text': answer.lower()})
            break
        qas.append({'question': question, 'id': qid, 'is_impossible': False, 'answers': answers})
        output.append({'context': context.lower(), 'qas': qas})
    return output
```

```python
qa_train = do_qa_train(train_array)
qa_train
```

```python
with open('datasets/train_transformer.json', 'w') as outfile:
    json.dump(qa_train, outfile)
```

### Test Data Preparation:

```python
def do_qa_test(test):
    output = []
    for line in test:
        context = line[1]
        qas = []
        question = line[-1]
        qid = line[0]
        if type(context)!= str or type(question)!= str:
            print(context, type(context))
            print(answer, type(answer))
            print(question, type(question))
            continue
        answers = []
        answers.append({'answer_start':1000000, 'text':'__None__'})
        qas.append({'question':question, 'id':qid, 'is_impossible':False, 'answers':answers})
        output.append({'context':context.lower(), 'qas':qas})
    return output
```
