 # Song Genre Classification

Streaming services have looked into means of categorizing music to allow for personalized recommendations. One method involves direct analysis of the raw audio information in a given song, scoring the raw data on a variety of metrics. This genre text data is compiled by a research group known as The Echo Nest. Our goal is to look through this dataset and classify songs based on their genre - all without listening to a single one ourselves.


# Data Exploration

1) danceability: [0–1] Describes how suitable a track is for dancing based on musical attributes including tempo, rhythm, stability, beat strength, and overall regularity.<br>
2) tempo — [0–300 BPM] The speed or pace of a given piece, as derived from the estimated average beat duration.<br>
3) liveness: [0–1] Detects the presence of an audience in the recording.<br>
4) instrumentalness: [0–1] Predicts whether a track contains no vocals (values above 0.5 represent instrumental tracks whereas rap songs would have a score close to 0).<br>
5) acousticness: [0–1] Confidence measure of whether the track is acoustic.<br>
6) speechiness: [0–1] Detects the presence of spoken words in a track (values above 0.66 describe tracks that are probably made entirely of spoken words, 0.33–0.66 describe tracks that may contain both music and speech, and values below 0.33 most likely represent music and other non-speech-like tracks).<br>
7) loudness: [-60–0 dB] The average volume across an entire track.<br>
8) energy: [0–1] Perceptual measure of intensity and activity. Energetic tracks feel fast, loud, and noisy (e.g. death metal: high energy, Bach prelude: low energy).<br>
9) valence: [0–1] Describes the musical positiveness conveyed by a track. Tracks with high valence sound more positive (e.g. happy, cheerful, euphoric), while tracks with low valence sound more negative (e.g. sad, depressed, angry).

## Outlier Identification

```python
def outlier_identify(column_list):
    total = 0
    for c in (column_list):
        q1, q3 = np.percentile(echo_tracks[c],[25,75])
        iqr = q3 - q1 #Interquartile range
        min_ = q1 - 1.5*iqr #The minimum value
        max_ = q3 + 1.5*iqr #The maximum value
        range_ = [min_, max_]
        min_, q1, q3, max_, range_
        outlier = echo_tracks[(echo_tracks[c]< min_) | (echo_tracks[c] > max_)]
        total += outlier.shape[0]
        print(c, 'has', outlier.shape[0], 'outliers outside of the range of:', range_)
        print(" ")
        print(outlier)
        print(120*'-')
    print('Total =', total, "rows (", total/echo_tracks.shape[0]*100, '% of full dataset)')
    
outlier_identify(['energy', 'loudness', 'speechiness'])
```

# Data Cleaning

## Outlier Removal
Data points outside of the IQR range/ 3 standard deviations will be removed to reduce noise to improve a model’s performance and prevent overfitting

```python
echo_tracks.shape
```

```python
for c in (['energy', 'loudness', 'speechiness']):
    q1, q3 = np.percentile(echo_tracks[c],[25,75])
    iqr = q3 - q1
    min_ = q1 - 1.5*iqr
    max_ = q3 + 1.5*iqr
    df_clean = echo_tracks.loc[(echo_tracks[c] >= min_) & (echo_tracks[c] <= max_)]
print(df_clean.shape)
```

# Data Split

Since there isn't any particularly strong correlation between the features, I will proceed to split the data into an array containing  features, and another containing the labels - the genre of the track. 

```python
from sklearn.model_selection import train_test_split

X = df_clean.drop(["genre","id", "song_name", "mode", "time_signature"], axis=1)
y = df_clean["genre"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10, stratify=y)
X_train.shape
```

# Data Standardization

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

# Principal Component Analysis (PCA)

Since there isn't any particular strong correlations between our features, I will use PCA to reduce the number of features, since it  is possible that the variance between genres can be explained by just a few features in the dataset.

Now that the data has been preprocessed, I can use PCA to determine by how much we can reduce the dimensionality of data, using scree-plots and cumulative explained ratio plots to find the number of components to use in further analyses.

```python
from sklearn.decomposition import PCA
pca_base = PCA()
pca_base.fit(scaled_X_train)
exp_variance = pca_base.explained_variance_ratio_

# plot the explained variance
fig, ax = plt.subplots()
ax.bar(range(pca_base.n_components_), exp_variance)
ax.set_xlabel('Principal Component #')
```

Some information is indeed lost in dimensionality reduction, but a fairly decent approximation can still be obtained with the tremendous benefit of being able to visualize the data. Moreover, the complexity of the problem can also be reduced by reducing the number of variables.

Scree-plot displays the number of components against the variance explained by each component, sorted in descending order of variance. However, there does not appear to be a clear elbow in this scree plot. I will use the cumulative explained variance plot to determine how many features are required to explain 85% of the variance. 

```python
cum_exp_variance = np.cumsum(exp_variance)

sns.set(style='whitegrid')
fig, ax = plt.subplots()
ax.plot(cum_exp_variance)
ax.axhline(y=0.85, linestyle='--', color='r', linewidth=4)
```

Looking at the cumulative explained variance plot, at 85% of cumulative explained variance, 6 features can explain the variance in our data

```python
from IPython.display import display

pca_df = pd.DataFrame()
pca_df['Cumulative Variance Ratio'] = cum_exp_variance
pca_df['Explained Variance Ratio'] = exp_variance
display(pca_df.head(10))
```

Each of the principal components explain a pretty considerable amount of variance. However, as 6 features can explain 85% of the variance, I will use 6 features

```python
pca = PCA(n_components=6, random_state=0)

X_train_pca = pca.fit_transform(scaled_X_train)
X_test_pca = pca.transform(scaled_X_test)
```

Now I can use the lower dimensional PCA projection of the data to classify songs into genres. 

# Decision Tree, Logistic Regression, KNN & Random Forest

I will use & compare the performance of 4 models. Since balancing the data has removed bias towards the more prevalent class, to get a good sense of how well the models are actually performing, I will use cross-validation to compare the models.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, KFold

models= {"DecisionTree": DecisionTreeClassifier(random_state=0), 
         "LogReg":LogisticRegression(random_state=0),
         "KNN":KNeighborsClassifier(n_neighbors=6),
         "Random Forest":RandomForestClassifier()}

def fit_and_score(models, X_train, X_test, y_train, y_test):
    np.random.seed(1)
    model_scores= {}
    results = []
    for name, model in models.items():
        model.fit(X_train, y_train)
        model_scores[name] = model.score(X_test, y_test)
        kf = KFold(n_splits=10, random_state=0, shuffle=True)
        cv_results = cross_val_score(model, X_train, y_train, cv=kf)
        results.append(cv_results)
    plt.boxplot(results, labels=models.keys())
    plt.show()
    return model_scores
```

```python
model_scores = fit_and_score(models=models, X_train=X_train_pca, X_test=X_test_pca, y_train=y_train, y_test=y_test)
model_scores
```

```python
model_compare = pd.DataFrame(model_scores, index=["Test Set Accuracy"])
model_compare.T.plot.barh(color=["lightblue"]);
```

Random Forest has the best performance, hence, it will be used for classification

# Hyperparameter Tweaking

I will perform hyperparameter tweaking to the Random Forest model to find the ideal parameters for models 

## 1st Round
I will start with RandomSearchCV to consider a wide range of values

I will be tuning these hyperparameters:<br>
- n_estimators: the number of “trees” in our Random Forest.<br>
- max_features: the number of features  provided to each tree in a random forest = at each split.<br>
- max_depth: the longest path between the root node and the leaf node = the max number of “splits” each tree can have.<br>
- min_samples_split: the minimum number of observations required in a any node before a node of a tree can split itself.<br>
- min_samples_leaf: the minimum number of observations that should be present in the leaf node after splitting a node.<br>
- bootstrap: whether to use bootstrapping (random sampling from the dataset with replacement) or not to provide data to each tree in the Random Forest.

```python
from sklearn.model_selection import RandomizedSearchCV

n_estimators = [int(x) for x in np.linspace(start=100, stop=1000, num=10)]
max_features = ['log2', 'sqrt']
max_depth = [int(x) for x in np.linspace(start=1, stop=15, num=15)]
min_samples_split = [int(x) for x in np.linspace(start=2, stop=50, num=10)]
min_samples_leaf = [int(x) for x in np.linspace(start=2, stop=50, num=10)]
bootstrap = [True, False]

params = {"n_estimators": n_estimators,"max_features": max_features, "max_depth": max_depth,  
          "min_samples_split": min_samples_split, "min_samples_leaf": min_samples_leaf, "bootstrap": bootstrap}

rs_rf = RandomizedSearchCV(RandomForestClassifier(), params, n_iter=100, cv=3, n_jobs=-1, verbose=True, random_state=0)
rs_rf.fit(X_train_pca, y_train)
```

With n_iter = 100 and cv = 3, 300 Random Forest models had been created, with randomly sampling combinations of the hyperparameters input above. The best performing model’s parameter can be checked. 

```python
print(rs_rf.best_params_)
```

```python
y_pred_rs_rf = rs_rf.predict(X_test_pca)
comparison = pd.DataFrame(data={"Actual":y_pred_rs_rf, "Prediction":y_test})
comparison.head(10)
```

```python
comparison["result"]=comparison["Actual"]==comparison["Prediction"]
comparison["result"].value_counts().plot(kind="barh", color=["lightblue"], title="Comparison")
```

```python
rs_rf_df = pd.DataFrame(rs_rf.cv_results_).sort_values('rank_test_score').reset_index(drop=True)
rs_rf_df = rs_rf_df.drop(['mean_fit_time', 'std_fit_time', 'mean_score_time','std_score_time', 
            'params', 
            'split0_test_score', 'split1_test_score', 
            'split2_test_score', 'std_test_score'], axis=1)
rs_rf_df.head(10)
```

Bar plots can be created for each hyperparameter on the x-axis, and the mean score of the models made at each value, to see which values for each hyperparameter were most successful on average.

```python
fig, axs = plt.subplots(ncols=3, nrows=2)
sns.set(style="whitegrid", color_codes=True, font_scale = 2)
fig.set_size_inches(30,25)

sns.barplot(x='param_n_estimators', y='mean_test_score', data=rs_rf_df, ax=axs[0,0], color='lightgrey')
#axs[0,0].set_ylim([.83,.93])
axs[0,0].set_title(label = 'n_estimators', size=30, weight='bold')

sns.barplot(x='param_min_samples_split', y='mean_test_score', data=rs_rf_df, ax=axs[0,1], color='coral')
#axs[0,1].set_ylim([.85,.93])
axs[0,1].set_title(label = 'min_samples_split', size=30, weight='bold')

sns.barplot(x='param_min_samples_leaf', y='mean_test_score', data=rs_rf_df, ax=axs[0,2], color='lightgreen')
#axs[0,2].set_ylim([.80,.93])
axs[0,2].set_title(label = 'min_samples_leaf', size=30, weight='bold')

sns.barplot(x='param_max_features', y='mean_test_score', data=rs_rf_df, ax=axs[1,0], color='wheat')
#axs[1,0].set_ylim([.88,.92])
axs[1,0].set_title(label = 'max_features', size=30, weight='bold')

sns.barplot(x='param_max_depth', y='mean_test_score', data=rs_rf_df, ax=axs[1,1], color='lightpink')
#axs[1,1].set_ylim([.80,.93])
axs[1,1].set_title(label = 'max_depth', size=30, weight='bold')

sns.barplot(x='param_bootstrap',y='mean_test_score', data=rs_rf_df, ax=axs[1,2], color='skyblue')
#axs[1,2].set_ylim([.88,.92])
axs[1,2].set_title(label = 'bootstrap', size=30, weight='bold')

plt.show()
```

## 2nd Round 

After using RandomSearchCV, GridSearchCV can be used with the average best performing ranges of each hyperparameter to perform a more refined search for the best hyperparameters. In GridSearchCV, every single combination of hyperparameter values is tried which takes much more computational power than RandomSearchCV, where we can directly control how many iterations we want to try. For demo purpose, due to resource constraint (CPU), I will stick to RandomSearchCV.

```python
n_estimators = [500,600,700,800,900]
max_features = ['sqrt']
max_depth = [12,13,14,15]
min_samples_split = [23,28,44]
min_samples_leaf = [2,18,23,34]
bootstrap = [False]

param_grid = {'n_estimators': n_estimators,'max_features': max_features,'max_depth': max_depth,
               'min_samples_split': min_samples_split,'min_samples_leaf': min_samples_leaf,'bootstrap': bootstrap}

rs_rf2 = RandomizedSearchCV(RandomForestClassifier(), param_grid, cv=3, verbose=True, n_jobs=-1)
rs_rf2.fit(X_train_pca, y_train)
```

# KMeans

In order to implement k-means clustering, I will use the 'elbow' method to select a number of clusters, k, which distinctly splits the data.

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans

scores_pca = pca.transform(scaled_X_train)
visualizer = KElbowVisualizer(KMeans(init='k-means++', random_state=42), k=(1,21), timings=False)
visualizer.fit(scores_pca)
visualizer.show()
n_clusters = visualizer.elbow_value_
print("Optimal number of clusters:", n_clusters)
```

The within-cluster sum of squares (WCSS) curve shows that the KneeLocator method has determined that 6 is the ideal number of clusters or distinct groupings to be used.<br><br>
Now, I can proceed to implement k-means clustering.

```python
kmeans_pca = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42)
kmeans_pca.fit(scores_pca)
```

```python
scaled_X_train_df = pd.DataFrame(scaled_X_train, columns=['acousticness','danceability','duration_ms','energy','instrumentalness','key','liveness','loudness','speechiness','valence','tempo'])

df_seg_pca_kmeans = pd.concat([scaled_X_train_df.reset_index(drop=True), pd.DataFrame(scores_pca)], axis=1)
df_seg_pca_kmeans.rename(columns={df_seg_pca_kmeans.columns[-1]: 'Component 6', df_seg_pca_kmeans.columns[-2]: 'Component 5', \
                                 df_seg_pca_kmeans.columns[-3]: 'Component 4', df_seg_pca_kmeans.columns[-4]: 'Component 3',\
                                 df_seg_pca_kmeans.columns[-5]: 'Component 2', df_seg_pca_kmeans.columns[-6]: 'Component 1'}, inplace=True)

df_seg_pca_kmeans['Cluster'] = kmeans_pca.labels_
df_seg_pca_kmeans.head()
```

What good do these cluster labels do? We can visualize the clusters on a 2D plane to ensure that there is some sensical separation of data based on the sources of signals that are encapsulated by the components.

```python
fig = plt.figure(figsize=(10, 8))
sns.scatterplot(x=df_seg_pca_kmeans['Component 2'], y=df_seg_pca_kmeans['Component 1'], hue=df_seg_pca_kmeans['Cluster'], palette=['tab:orange','tab:green','tab:red','tab:purple','goldenrod','tab:cyan'])
#Need to define x & y directly iside the scatterplot for it to work!!!
plt.title('Clusters by PCA Components', fontsize=18)
plt.xlabel("Component 2", fontsize=18)
plt.ylabel("Component 1", fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

In the figure above, there are clear groupings as indicated by the concentrated areas, for example, Cluster 0 is most similar to Clusters 5, but overall, the clusters overlap, hence, we cannot rule out the possibility of noises. Additionally, this is only a 2D visualization of a 6-dimensional vector space, so more analysis is needed in order to determine the efficacy of the k-means algorithm.<br><br>
This clustering can be further expanded in the future to cater to use case such as building a personal playlist.
